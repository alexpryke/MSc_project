```python
import pandas as pd                        
from pytrends.request import TrendReq
pytrend = TrendReq()
```

```python
# american extreme weather tweets over time
pytrend.build_payload(kw_list=['Climate Change'], geo = 'US-CA')
df = pytrend.interest_over_time()
df.head(10)
print(df)
df.reset_index().plot(x='date', y='Climate Change', figsize=(120, 10), kind ='bar')
```

```python
# american extreme weather tweets over time
pytrend.build_payload(kw_list=['Global Warming'], geo = 'US-CA')
df = pytrend.interest_over_time()
df.head(10)
print(df)
df.reset_index().plot(x='date', y='Global Warming', figsize=(120, 10), kind ='bar')
```

```python
# american extreme weather tweets over time
pytrend.build_payload(kw_list=['Global Warming', 'Climate Change'], geo = 'US-CA')
df = pytrend.interest_over_time()
df.head(10)
print(df)
df.reset_index().plot(x='date', y=['Global Warming', 'Climate Change'], figsize=(120, 10), kind ='bar')
```
